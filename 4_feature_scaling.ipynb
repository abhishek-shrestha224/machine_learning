```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("data/500hits.csv", encoding="latin-1")
df.sample(n=10)
```

```python
df.drop(columns=['PLAYER','HOF'], inplace=True)
```

```python
df.info()
```

```python
df.describe().round(3)
```

```python
X1= df
X2= df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
scaleStandard = StandardScaler()
```

```python
scaled_X1= scaleStandard.fit_transform(X1)
```

```python
scaled_X1=pd.DataFrame(scaled_X1, columns = X1.columns)
```

```python
scaled_X1.describe().round(3)
```

```python
scaleMinMax = MinMaxScaler(feature_range=(-1,1))
```

```python
scaled_X2=scaleMinMax.fit_transform(X2)
```

```python
scaled_X2=pd.DataFrame(scaled_X2, columns = X2.columns)
```

```python
scaled_X2.describe().round(3)
```

```python
n_cols = len(df.columns)
fig, axs = plt.subplots(7, 2, figsize=(20, 25))

axs = axs.flatten()

for i, column in enumerate(df.columns):
    sns.histplot(df[column], kde=True, ax=axs[i])
    axs[i].set_title(f"Histogram of {column}")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
n_cols = len(scaled_X1.columns)
fig, axs = plt.subplots(7, 2, figsize=(20, 25))

axs = axs.flatten()

for i, column in enumerate(scaled_X1.columns):
    sns.histplot(scaled_X1[column], kde=True, ax=axs[i])
    axs[i].set_title(f"Histogram of {column}")
plt.grid(True)
plt.tight_layout()
plt.show()
```
