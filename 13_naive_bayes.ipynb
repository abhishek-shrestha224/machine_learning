```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
merged["day_of_week"].unique()
```

```python
days = ["Sunday", "Monday", "Tuesday", "Thursday", "Friday", "Saturday"]
```

```python
enc = OrdinalEncoder(categories=[days])
```
