```python
import pandas as pd
```

```python
dat = {
    'sales_amount': [
        1200, 1500, 1100, 1300, 1400, 1250, 1450, 1550, 1350, 1600,
        1650, 1700, 1750, 1800, 1850, 1900, 1950, 2000, 2050, 2100
    ],
    'size': [
        'md', 'lg', 'sm', 'md', 'sm', 'md', 'sm', 'xl', 'md', 'sm',
        'lg', 'sm', 'xl', 'md', 'lg', 'lg', 'md', 'lg', 'lg', 'sm'
    ],
    'city_name': [
        'Kathmandu', 'Bhaktapur', 'Pokhara', 'Lalitpur', 'Chitwan', 'Dhulikhel', 'Kathmandu', 'Bhaktapur', 'Pokhara', 'Lalitpur',
        'Chitwan', 'Dhulikhel', 'Kathmandu', 'Bhaktapur', 'Pokhara', 'Lalitpur', 'Chitwan', 'Dhulikhel', 'Kathmandu', 'Bhaktapur'
    ]
}
```

```python
df = pd.DataFrame(data=dat)
df.head(n=20)
```

```python
from sklearn.preprocessing import  OneHotEncoder, OrdinalEncoder
```

```python
ohe = OneHotEncoder(sparse_output=False).set_output(
    transform="pandas"
)
```

```python
ode = OrdinalEncoder(categories=[['sm','md','lg','xl']])
```

```python
from sklearn.compose import ColumnTransformer
```

```python
ct = ColumnTransformer(
    transformers=[
        ('ohe', ohe, ['city_name']),  # ('name', transformer, columns)
        ('ode', ode, ['size'])
    ],
    remainder='passthrough'
)
```

```python
ct.set_output(transform="pandas")
```

```python
df_final = ct.fit_transform(df)
df_final
```
