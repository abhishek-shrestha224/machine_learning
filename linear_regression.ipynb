```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
dat = {
    "area": [2600, 3000, 3200, 3600, 4000],
    "bedrooms": [3, 4, None, 3, 5],
    "age": [20, 15, 18, 30, 8],
    "price": [500_000, 565_000, 610_000, 595_000, 760_000],
}
```

```python
df = pd.DataFrame(dat)
```

```python
df
```

```python
import math

md_bedrooms = math.floor(df["bedrooms"].median())
print(md_bedrooms)
```

```python
df["bedrooms"] = df["bedrooms"].fillna(md_bedrooms)
```

```python
df
```

```python
reg = LinearRegression()
reg.fit(X=df[["area", "bedrooms", "age"]], y=df["price"])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000, 3, 40]])
```
